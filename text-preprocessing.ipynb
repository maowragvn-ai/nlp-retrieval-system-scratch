### **1.1 Normal**

```python
def to_lowercase(text: str):
  return text.lower()

def to_uppercase(text: str):
  return text.upper()

def remove_html_tag(text: str):
  return re.sub(r'<[^>]*>', '', text)

def remove_url(text: str):
  return re.sub(r'http\S+', '', text)

def remove_punctuation_character(text: str):
  punctuations = string.punctuation
  return text.translate(str.maketrans('', '', punctuations))

def remove_special_character(text: str):
  return re.sub(r'[^a-zA-Z0-9\s]', '', text)

def remove_number(text: str):
  return re.sub(r'\d+', '', text)

def reduce_whitespace(text: str):
  return re.sub(r'\s+', ' ', text)

def remove_emoji(string):
  emoji_pattern = re.compile("[" u"\U0001F600-\U0001F64F" # emoticons
                             u"\U0001F300-\U0001F5FF" # symbols & pictographs
                             u"\U0001F680-\U0001F6FF" # transport & map symbols
                             u"\U0001F1E0-\U0001F1FF" # flags (iOS)
                             u"\U00002500-\U00002BEF" # chinese char
                             u"\U00002702-\U000027B0"
                             u"\U00002702-\U000027B0"
                             u"\U000024C2-\U0001F251"
                             u"\U0001f926-\U0001f937"
                             u"\U00010000-\U0010ffff"
                             u"\u2640-\u2642"
                             u"\u2600-\u2B55"
                             u"\u200d"
                             u"\u23cf"
                             u"\u23e9"
                             u"\u231a"
                             u"\ufe0f" # dingbats
                             u"\u3030" "]+", flags=re.UNICODE)
  return emoji_pattern.sub(r'', string)
```

### **1.2 Stop/Slang Words remove**

```python
def chat_slang_translator_pd(text: str):
    # File path which consists of Abbreviations.
    fileName = os.path.join(ss_words_path, "slang.txt")

    # Read the slang file into a pandas DataFrame
    try:
        slang_df = pd.read_csv(fileName, sep='=', header=None, names=['abbreviation', 'phrase'])
    except FileNotFoundError:
        print(f"Error: Slang file not found at {fileName}")
        return text # Return original text if file not found

    text_list = text.split(" ")
    translated_text_list = []

    for word in text_list:
        # Removing Special Characters.
        cleaned_word = re.sub('[^a-zA-Z0-9-_.]', '', word)

        # Check if selected word matches abbreviations in the DataFrame
        match = slang_df[slang_df['abbreviation'] == cleaned_word.upper()]

        if not match.empty:
            # If match found replace it with its appropriate phrase
            translated_text_list.append(match.iloc[0]['phrase'])
        else:
            # If no match, keep the original word
            translated_text_list.append(word)

    # Joining the words back into a string
    return ' '.join(translated_text_list)
```

### **1.5 Stemming/Lemmatization**

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
def stemming_word(text: str, stem_type: str = "porter"):
  if stem_type == "snowball":
    snowball_stemmer = SnowballStemmer('english')
    return ' '.join([snowball_stemmer.stem(word) for word in text.split()])
  else:
    porter_stemmer = PorterStemmer()
    return ' '.join([porter_stemmer.stem(word) for word in text.split()])
```

```python
print(stemming_word("Running flying buying bought went flies", "snowball"))
```

```python
def lemmatization_word(text: str):
  lemmatizer = WordNetLemmatizer()
  return ' '.join([lemmatizer.lemmatize(word,pos='v') for word in text.split()])
```

```python
print(lemmatization_word("Running flying buying bought went flies"))
```
